### 3. Limpieza de datos

**Reemplazar valores no válidos**

```python
df.fillna(0,inplace=True)
df
```

**Corregir tipo de datos (Las medallas son datos de tipo float y queremos que sean de tipo int)**

```python
df['Oro']=df['Oro'].astype(int)
df['Plata']=df['Plata'].astype(int)
df['Bronce']=df['Bronce'].astype(int)
df
```

#### 4. Análisis de Medallas de Oro por País

#### 5. Análisis de Medallas Totales por País

```python
filtro=df['Total']>10
mas_de_10_medallas=df[filtro]
mas_de_10_medallas.sort_values('Total', ascending=False)
```
